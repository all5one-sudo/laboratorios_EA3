```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('E:\Documentos\Informes Latex\TP1_EA3\calculo.xlsx', engine='openpyxl')

N = df['N']
l_D = df['l/D']

plt.figure(figsize=(10, 6))
plt.plot(l_D, N)

plt.xlabel('l/D')
plt.ylabel('N (cantidad de espiras)')

x_value = 1.3
y_value = 10

plt.scatter(x_value, y_value, color='red')
plt.text(x_value, y_value, f'({x_value}, {y_value} espiras)', fontsize=12, ha='right')
plt.grid(True)
plt.savefig('figura10.pdf')

plt.show()
```

```python
Rp = df['Rp [Ohm]'] 
Rt = df['Rt [Ohm]'] 
Rgp = df['Rgp [Ohm]'] 
Rlp = df['RLp [Ohm]']

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(l_D, Rp, color='blue')
axs[0, 0].set_title('Resistencia de p칠rdidas de la bobina')
axs[0, 0].set_xlabel('l/D')
axs[0, 0].set_ylabel(r'$Rp [\Omega]$')
axs[0, 0].grid(True)

axs[0, 1].plot(l_D, Rt, color='green')
axs[0, 1].set_title('Resistencia total')
axs[0, 1].set_xlabel('l/D')
axs[0, 1].grid(True)

axs[1, 0].plot(l_D, Rgp, color='red')
axs[1, 0].set_title("Resistencia del generador reflejada")
axs[1, 0].set_xlabel('l/D')
axs[1, 0].grid(True)

axs[1, 1].plot(l_D, Rlp, color='purple')
axs[1, 1].set_title("Resistencia de carga reflejada")
axs[1, 1].set_xlabel('l/D')
axs[1, 1].grid(True)

axs[0, 1].set_ylabel(r'$Rt [\Omega]$')
axs[1, 0].set_ylabel(r"$Rg' [\Omega]$")
axs[1, 1].set_ylabel(r"$Rl' [\Omega]$")

fig.tight_layout()

plt.savefig('figura12.pdf')

plt.show()
```

```python
import numpy as np

data = np.genfromtxt('E:\\Documentos\\Informes Latex\\TP1_EA3\\simulacion.txt', delimiter='\t', dtype=None, names=('F', 'V'), encoding=None)

data_valid = [v for v in data['V'] if len(v.split(',')) > 1 and v.split(',')[0][1:] and v.split(',')[1][:-1]]
magnitudes = [float(v.split(',')[0][1:]) for v in data_valid]
phases = [float(v.split(',')[1][:-1]) for v in data_valid]

fig, axs = plt.subplots(2, 1)

axs[0].plot(data['F'][:len(data_valid)], magnitudes, color='blue')
axs[0].set_title('Magnitud')
axs[0].set_xlabel('Frecuencia')
axs[0].set_ylabel('Magnitud [dB]')
axs[0].set_xticks([data['F'][1], data['F'][len(data_valid)//2+1], data['F'][len(data_valid)-1]])  # Mostrar solo las graduaciones seleccionadas
axs[0].set_xticklabels([data['F'][1], data['F'][len(data_valid)//2+1], data['F'][len(data_valid)-1]])  # Establecer las etiquetas de las graduaciones
axs[0].grid(True)
axs[0].set_ylim(-5, 10)

axs[1].plot(data['F'][:len(data_valid)], phases, color='green')
axs[1].set_title('Fase')
axs[1].set_xlabel('Frecuencia')
axs[1].set_ylabel('Fase [춿]')
axs[1].set_xticks([data['F'][1], data['F'][len(data_valid)//2+1], data['F'][len(data_valid)-1]])  # Mostrar solo las graduaciones seleccionadas
axs[1].set_xticklabels([data['F'][1], data['F'][len(data_valid)//2+1], data['F'][len(data_valid)-1]])  # Establecer las etiquetas de las graduaciones
axs[1].grid(True)
fig.tight_layout()

plt.savefig('figura16.pdf')
plt.show()
```
