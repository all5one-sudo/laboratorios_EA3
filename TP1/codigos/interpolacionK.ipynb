# Interpolación polinómica Curva K

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = np.array([0.75,0.8,0.85,0.9,0.95,1,1.05,1.1,1.15,1.2,1.25])
y = np.array([4.6,5,5.5,6,6.4,6.6,7,7.2,7.5,7.8,8,])

coeffs = np.polyfit(x, y, 3)
polynomial = np.poly1d(coeffs)

x_plot = np.linspace(min(x), max(x), 1000)
y_plot = polynomial(x_plot)

plt.figure(figsize=(10, 6))
plt.semilogx(x, y, 'ro', label='Puntos originales')
plt.semilogx(x_plot, y_plot, label='Interpolación polinómica')
plt.legend(loc = 'lower right')
plt.xlabel('l/D')
plt.ylabel('Factor K')
plt.grid(True)

x_symbol = sp.symbols('x')
polynomial_sympy = sum(sp.S("{:6.2f}".format(v))*x_symbol**i for i, v in enumerate(coeffs[::-1]))
equation = "y = " + str(polynomial_sympy)
plt.text(0.05, 0.95, equation, transform=plt.gca().transAxes, fontsize=14,
         verticalalignment='top')

plt.title('Interpolación polinómica de grado 3')
plt.savefig('figura11.pdf')

plt.show()
```
